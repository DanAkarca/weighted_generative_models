```python
# set generative model hyperparameters
m = int(np.size(np.where(dk_consensus.flat))/2) # number of connections
eta = np.array([-2]) # eta parameter
gamma = np.array([.5]) # gamma parameter
model_type = 'matching' # generative model
model_var = 'powerlaw' # wiring equation
epsilon = 1e-6
nnode = len(dk_consensus)
```

```python
# set a seed network to grow the topology from
Aseed = np.zeros((nnode,nnode)) # seed the seed as a random edge that exists
```

```python
# compute a gradient
def matrix_communicability(x):
    y = jscipy.linalg.expm(x)
    return jnp.sum(y)
```

```python
# set input
x = jnp.array(dk_consensus)
# set loss function
loss = matrix_communicability
# compute the gradient
grad_loss = grad(loss)
# print the output
grad_loss(x)
```
